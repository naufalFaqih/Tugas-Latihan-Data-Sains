```python
import pandas as pd
import glob
import os

# Tentukan direktori tempat file Excel disimpan
# Ubah path sesuai dengan lokasi file Excel Anda
directory = directory = r"C:\Users\dzak\Documents\SEMESTER 6\MACHINE LEARNING\TUGAS & LATIHAN DATA SAINS\coba"  # Ganti dengan path folder yang berisi file Excel

# Kolom-kolom yang ingin diambil
selected_columns = ["nama", "usia", "tekanan_darah", "kadar_gula"]

# Cari semua file Excel di direktori tersebut
excel_files = glob.glob(os.path.join(directory, "*.xlsx"))

# Jika tidak ada file yang ditemukan
if len(excel_files) == 0:
    print("Tidak ada file Excel yang ditemukan di direktori tersebut.")
else:
    print(f"Ditemukan {len(excel_files)} file Excel.")
    
    # Buat list untuk menyimpan dataframe dari setiap file
    all_dataframes = []
    
    # Loop melalui setiap file Excel
    for file in excel_files:
        try:
            # Baca file Excel
            df = pd.read_excel(file)
            
            # Periksa apakah semua kolom yang diinginkan ada dalam file
            missing_columns = [col for col in selected_columns if col not in df.columns]
            if missing_columns:
                print(f"Warning: File {os.path.basename(file)} tidak memiliki kolom: {missing_columns}")
                # Opsional: tambahkan kolom yang hilang dengan nilai NaN
                for col in missing_columns:
                    df[col] = pd.NA
            
            # Ambil hanya kolom yang diinginkan
            df_selected = df[selected_columns].copy()
            
            # Tambahkan informasi nama file sebagai kolom baru (opsional)
            df_selected['source_file'] = os.path.basename(file)
            
            # Tambahkan dataframe ke list
            all_dataframes.append(df_selected)
            
            # Tampilkan informasi file yang sedang diproses
            print(f"Membaca file: {os.path.basename(file)}, jumlah baris: {len(df_selected)}")
            
        except Exception as e:
            print(f"Error membaca file {os.path.basename(file)}: {str(e)}")
    
    # Gabungkan semua dataframe
    if all_dataframes:
        # Gunakan axis=0 untuk menumpuk dataframe secara vertikal (menambah baris)
        combined_df = pd.concat(all_dataframes, axis=0, ignore_index=True)
        
        # Tampilkan informasi hasil penggabungan
        print(f"\nHasil penggabungan: {len(combined_df)} baris total")
        
        # Cek kolom-kolom yang ada
        print(f"Kolom dalam dataset gabungan: {combined_df.columns.tolist()}")
        
        # Simpan hasil ke file Excel baru
        output_path = os.path.join(directory, "combined_selected_columns.xlsx")
        combined_df.to_excel(output_path, index=False)
        print(f"\nFile hasil penggabungan disimpan di: {output_path}")
        
        # Tampilkan beberapa baris awal dari hasil penggabungan
        print("\nPreview hasil penggabungan:")
        display(combined_df.head())
    else:
        print("Tidak ada data valid untuk digabungkan.")
```
